```python
#Import all libraries and layers, prep notebook
import pickle
import datetime
import random
import csv
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import image
from sklearn.utils import shuffle
from skimage.color import rgb2gray

from tensorflow.keras import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, AveragePooling2D, Dropout, BatchNormalization, Activation
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical

%matplotlib inline
random.seed(datetime.datetime.now())
```

```python
#Load the dataset
training_file = '../input/german_traffic_sign_dataset/train.p'
validation_file='../input/german_traffic_sign_dataset/valid.p'
testing_file = '../input/german_traffic_sign_dataset/test.p'

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
with open(validation_file, mode='rb') as f:
    valid = pickle.load(f)
with open(testing_file, mode='rb') as f:
    test = pickle.load(f)
    
x_train, y_train = train['features'], train['labels']
x_valid, y_valid = valid['features'], valid['labels']
x_test, y_test = test['features'], test['labels']

#Combine the train, validation, and test set
X = np.concatenate((x_train, x_valid, x_test))
Y = np.concatenate((y_train, y_valid, y_test))

#Shuffle all images
X, Y = shuffle(X, Y)

train_index = int(0.7*len(X))
valid_index = int(0.9*len(X))

#Re-split after shuffling
x_train = X[:train_index]
x_valid = X[train_index:valid_index]
x_test = X[valid_index:]

y_train = Y[:train_index]
y_valid = Y[train_index:valid_index]
y_test = Y[valid_index:]
```

```python
n_train = x_train.shape[0]

n_validation = x_valid.shape[0]

n_test = x_test.shape[0]

image_shape = x_train.shape[1:]

n_classes = np.max(y_train) + 1

print("Number of training examples =", n_train)
print("Number of testing examples =", n_test)
print("Image data shape =", image_shape)
print("Number of classes =", n_classes)
```

```python
unique_train, counts_train = np.unique(y_train, return_counts=True)
unique_valid, counts_valid = np.unique(y_train, return_counts=True)
unique_test, counts_test = np.unique(y_train, return_counts=True)
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10,10))
ax1.bar(unique_train, counts_train)
ax1.set_title("Number of images per label in training set")
ax2.bar(unique_valid, counts_valid)
ax2.set_title("Number of images per label in validation set")
ax3.bar(unique_test, counts_test)
ax3.set_title("Number of images per label in test set")

fig.tight_layout()
plt.show()
```

```python
rand1 = random.randint(0, len(x_train))
rand2 = random.randint(0, len(x_valid))
rand3 = random.randint(0, len(x_test))

fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10,10))
ax1.imshow(x_train[rand1])
ax1.set_title("Example image from training set \n label: " + str(y_train[rand1]))
ax2.imshow(x_valid[rand2])
ax2.set_title("Example image from validation set \n label: " + str(y_valid[rand2]))
ax3.imshow(x_test[rand3])
ax3.set_title("Example image from test sett \n label: " + str(y_test[rand3]))
fig.tight_layout()
plt.show()
```

```python
#One hot encode labels
y_train = to_categorical(y_train, n_classes)
y_valid = to_categorical(y_valid, n_classes)
y_test = to_categorical(y_test, n_classes)

#Grayscal
x_train = np.sum(x_train/3, axis=3, keepdims=True)
x_valid = np.sum(x_valid/3, axis=3, keepdims=True)
x_test = np.sum(x_test/3, axis=3, keepdims=True)
x_train, y_train = shuffle(x_train, y_train)

#Normalize
x_train = (x_train - 128) / 128
x_valid = (x_valid - 128) / 128
x_test = (x_test - 128) / 128

#Reset image shape after grayscaling
image_shape = x_train.shape[1:]

#Double check normalization
print(np.max(x_valid))
```

```python
#Hyperparameters
batch_size = 3000
epochs = 200
learn_rate = 0.001
lam = 0.0001
```

```python
#Feed in batches
datagen = ImageDataGenerator()
datagen.fit(x_train)
```

```python
#Create model
model = Sequential()
model.add(Conv2D(6,
                 kernel_size=(3, 3),
                 strides=(1, 1),
                 activation='tanh', 
                 input_shape=image_shape, 
                 padding="same", 
                 kernel_regularizer=tf.keras.regularizers.l1(lam),
                 activity_regularizer=tf.keras.regularizers.l2(lam)
                ))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'))
model.add(Conv2D(16, 
                 kernel_size=(3, 3),
                 strides=(1, 1),
                 activation='tanh',
                 padding='same', 
                 kernel_regularizer=tf.keras.regularizers.l1(lam),
                 activity_regularizer=tf.keras.regularizers.l2(lam)
                ))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'))
model.add(Flatten())
model.add(Dense(800, activation='tanh'))
model.add(Dense(600, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(400, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(120, activation='tanh'))
model.add(Dropout(0.4))
model.add(Dense(84, activation='tanh'))
model.add(Dropout(0.4))
model.add(Dense(n_classes, activation='softmax'))

opt = keras.optimizers.Adam(learning_rate=learn_rate)
model.compile(optimizer=opt,
            loss=categorical_crossentropy,
            metrics=['accuracy'])
```

```python
#Display summary of model
model.summary()
```

```python
#Create callbacks for early stopping and tensorboard
log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
# Specify the callback object
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2)
```

```python
#Train the model
model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
          steps_per_epoch=len(x_train) / batch_size, epochs=epochs, 
          validation_data=(x_valid, y_valid),
          callbacks=[callback, tensorboard_callback])
```

```python
#Evaluate test set
loss, acc = model.evaluate(x_test, y_test)
```
