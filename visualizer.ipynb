```python
import csv
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

```python
with open('signnames.csv', mode='r') as infile:
    reader = csv.reader(infile)
    mydict = {rows[0]:rows[1] for rows in reader}
```

```python
training_file = '../input/german_traffic_sign_dataset/train.p'
validation_file='../input/german_traffic_sign_dataset/valid.p'
testing_file = '../input/german_traffic_sign_dataset/test.p'

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
    
x_train, y_train = train['features'], train['labels']
```

```python
fig=plt.figure(figsize=(100, 300))
fig.tight_layout()
num_classes = np.max(y_train)
for i in range(num_classes):
    fig.add_subplot(num_classes + 1, 1, i + 1)
    plt.imshow(x_train[np.where(y_train == i)[0][0]])
    plt.title(str(i) + " : " + mydict[str(i)])
plt.show()
```
